```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame creation (your actual data should replace this)
data = {
    'dataset': ['gsm8k']*5 + ['mt-bench']*5,
    'batch_size': [1, 1, 1, 1, 1, 128, 128, 128, 128, 128],
    'draft_size': [2, 3, 4, 5, 6, 4, 5, 6, 7, 8],
    'E(min(p q))': [0.665, 0.614, 0.597, 0.564, 0.546, 0.617, 0.583, 0.554, 0.570, 0.597],
    'E(min(1 p/q))': [0.824, 0.783, 0.781, 0.716, 0.709, 0.774, 0.705, 0.698, 0.693, 0.734]
}
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))

# Filter by dataset
gsm8k_df = df[df['dataset'] == 'gsm8k']
mt_bench_df = df[df['dataset'] == 'mt-bench']

# Plot gsm8k values
plt.scatter(gsm8k_df['draft_size'], gsm8k_df['E(min(1 p/q))'], color='blue', marker='o', label='gsm8k: Beta')
plt.scatter(gsm8k_df['draft_size'], gsm8k_df['E(min(p q))'], color='green', marker='o', label='gsm8k: Alpha')

# Plot mt-bench values
plt.scatter(mt_bench_df['draft_size'], mt_bench_df['E(min(1 p/q))'], color='blue', marker='s', label='mt-bench: Beta')
plt.scatter(mt_bench_df['draft_size'], mt_bench_df['E(min(p q))'], color='green', marker='s', label='mt-bench: Alpha')

# Adding titles and labels
plt.title('E(min(1, p/q)) vs. E(min(p, q)) by Draft Size and Dataset')
plt.xlabel('Draft Size')
plt.ylabel('Values')
plt.xticks(df['draft_size'].unique())  # Ensure we have a tick for each unique draft size
plt.legend()

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Plot beta values
plt.scatter(df['draft_size'], df['E(min(1 p/q))'], color='blue', label='Beta')

# Plot accept_rate values
plt.scatter(df['draft_size'], df['accept_rate'], color='red', label='Accept Rate')

# Adding titles and labels
plt.title('E(min(1 p/q) vs. Accept Rate by Draft Size')
plt.xlabel('Draft Size')
plt.ylabel('Values')
plt.xticks(df['draft_size'])  # Ensure we have a tick for each draft size
plt.legend()

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Plot beta values
plt.scatter(df['draft_size'], df['E(min(1 p/q))'], color='blue', label='E(min(1 p/q))')

# Plot alpha values
plt.scatter(df['draft_size'], df['E(min(p q))'], color='green', label='E(min(p q))')

# Adding titles and labels
plt.title('E(min(1, p/q)) vs. E(min(p, q)) by Draft Size')
plt.xlabel('Draft Size')
plt.ylabel('Values')
plt.xticks(df['draft_size'])  # Ensure we have a tick for each draft size
plt.legend()

# Show the plot
plt.show()
```
